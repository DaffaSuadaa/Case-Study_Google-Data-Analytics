### Cleaning Data

```python
q2_2019.rename(columns={
    '01 - Rental Details Rental ID': 'ride_id',
    '01 - Rental Details Bike ID': 'rideable_type',
    '01 - Rental Details Local Start Time': 'started_at',
    '01 - Rental Details Local End Time': 'ended_at',
    '03 - Rental Start Station Name': 'start_station_name',
    '03 - Rental Start Station ID': 'start_station_id',
    '02 - Rental End Station Name': 'end_station_name',
    '02 - Rental End Station ID': 'end_station_id',
    'User Type': 'member_casual'
}, inplace=True)

q3_2019.rename(columns={
    'trip_id': 'ride_id',
    'bikeid': 'rideable_type',
    'start_time': 'started_at',
    'end_time': 'ended_at',
    'from_station_name': 'start_station_name',
    'from_station_id': 'start_station_id',
    'to_station_name': 'end_station_name',
    'to_station_id': 'end_station_id',
    'usertype': 'member_casual'
}, inplace=True)

q4_2019.rename(columns={
    'trip_id': 'ride_id',
    'bikeid': 'rideable_type',
    'start_time': 'started_at',
    'end_time': 'ended_at',
    'from_station_name': 'start_station_name',
    'from_station_id': 'start_station_id',
    'to_station_name': 'end_station_name',
    'to_station_id': 'end_station_id',
    'usertype': 'member_casual'
}, inplace=True)
```

```python
q2_2019['ride_id'] = q2_2019['ride_id'].astype('object')
q2_2019['rideable_type'] = q2_2019['rideable_type'].astype('object')

q3_2019['ride_id'] = q3_2019['ride_id'].astype('object')
q3_2019['rideable_type'] = q3_2019['rideable_type'].astype('object')

q4_2019['ride_id'] = q4_2019['ride_id'].astype('object')
q4_2019['rideable_type'] = q4_2019['rideable_type'].astype('object')
```

```python
all_bike_df = pd.concat([q2_2019, q3_2019, q4_2019, q1_2020], ignore_index=True)
```

```python
all_bike_df.info()
```

```python
all_bike_df.isna().sum()
```

```python
drop_col = ['01 - Rental Details Duration In Seconds Uncapped', 'Member Gender', '05 - Member Details Member Birthday Year', 'tripduration', 'gender', 'birthyear', 'start_lat', 'start_lng', 'end_lat', 'end_lng']

all_bike_df.drop(columns=drop_col, inplace=True)

all_bike_df.head(5)
```

```python
all_bike_df.isna().sum()
```

```python
rows_with_missing = all_bike_df[all_bike_df.isnull().any(axis=1)]

print("Baris dengan data kosong:")
print(rows_with_missing)
```

```python
all_bike_df = all_bike_df.dropna(subset=['end_station_id'])

print("Data setelah menghapus nilai null:")
all_bike_df.isna().sum()
```

```python
all_bike_df['member_casual'].value_counts()
```

```python
all_bike_df['member_casual'] = all_bike_df['member_casual'].replace({
    'Subscriber': 'member',
    'Customer': 'casual'
})

all_bike_df['member_casual'].value_counts()
```

```python
datetime_columns = ["started_at", "ended_at"]

for column in datetime_columns:
  all_bike_df[column] = pd.to_datetime(all_bike_df[column])

all_bike_df.info()
```

```python
all_bike_df['year'] = all_bike_df['started_at'].dt.year
all_bike_df['month'] = all_bike_df['started_at'].dt.month
all_bike_df['day'] = all_bike_df['started_at'].dt.day
all_bike_df['day_of_week'] = all_bike_df['started_at'].dt.day_name()

all_bike_df.head()
```

```python
all_bike_df['day_of_week'].value_counts()
```

```python
all_bike_df['ride_length'] = all_bike_df['ended_at'] - all_bike_df['started_at']

all_bike_df['ride_in_minute'] = all_bike_df['ride_length'].dt.total_seconds()
all_bike_df.head()
```

```python
filtered_data = all_bike_df[(all_bike_df['start_station_name'] == 'HQ QR') | (all_bike_df['ride_in_minute'] < 0)]

filtered_data['ride_id'].count()
```

```python
all_bike_df = all_bike_df[~((all_bike_df['start_station_name'] == 'HQ QR') | (all_bike_df['ride_in_minute'] < 0))]

all_bike_df.describe()
```

**Insight:**
* Semua column dan tipe data telah sama dengan bentuk dataframe q1_2020.
* Setelah semua dataframe digabungkan dan menghapus beberapa column yang tidak diperlukan.
* Membersihkan data kosong.
* Menentukan penggunaan data yang konsisten pada column member_casual (Customer dan member >> member dan casual).
* Mengganti tipedata start dan end dengan tipe data datetime
* Menghapus data dummy.

## Exploratory Data Analysis (EDA)

```python
all_bike_df.groupby('member_casual').agg({'ride_in_minute' : 'count'})
```

```python
all_data_mean = all_bike_df.groupby(['member_casual','day_of_week']).agg({'ride_in_minute' : 'mean'})
print(all_data_mean)
```

```python
member_data = all_bike_df[all_bike_df['member_casual']  == 'member']

popular_start_station_member = member_data.groupby('start_station_name').agg({'member_casual' : 'count'}).sort_values(by= 'member_casual', ascending=False)

popular_start_station_member.head(7)
```

```python
casual_data = all_bike_df[all_bike_df['member_casual']  == 'casual']

popular_start_station_casual = casual_data.groupby('start_station_name').agg({'member_casual' : 'count'}).sort_values(by='member_casual', ascending=False)

popular_start_station_casual.head(7)
```

## Visualization & Explanatory Analysis

```python
all_bike_df['day_of_week'] = pd.Categorical(all_bike_df['day_of_week'],
                                    categories=['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'],
                                    ordered=True)

all_data_mean = all_bike_df.groupby(['member_casual','day_of_week'], observed=True).agg({'ride_in_minute' : 'count'})
plt.figure(figsize= (10,8))

sns.barplot(data=all_data_mean , y="ride_in_minute", x="day_of_week", hue="member_casual")

plt.show()
```

```python
all_data_mean = all_bike_df.groupby(['member_casual','day_of_week'], observed=True).agg({'ride_in_minute' : 'mean'})
plt.figure(figsize= (10,8))

sns.barplot(data=all_data_mean , y="ride_in_minute", x="day_of_week", hue="member_casual")

plt.show()
```
